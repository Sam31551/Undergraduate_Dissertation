# Read in Data

```python
import pandas as pd
data = pd.read_csv('FE_Data.csv', index_col=0)
data
```

```python
scalable_data = data[['goals', 'gd', 'opp_gd', 'avg_starter_age', 'total_starter_value', 'win%' ]]
scalable_data
```

```python
unscalable_data = data[['country', 'team', 'season', 'md', 'outcome', 'Cluster', 'win']]
unscalable_data 
```

```python
#Use MinMax Scaler as upper and lower bounds are known 
from sklearn.preprocessing import MinMaxScaler
minmaxsc = MinMaxScaler()
scaled_data = pd.DataFrame(minmaxsc.fit_transform(scalable_data.values), columns=scalable_data.columns, index=scalable_data.index)
scaled_data.head()
```

```python
# Place the DataFrames side by side
horizontal_stack = pd.concat([ unscalable_data, scaled_data, ], axis=1)
horizontal_stack.head()
```

```python
scaled_data = horizontal_stack
scaled_data
```

```python
scaled_data.to_csv('scaled_data.csv', index=None)
```
