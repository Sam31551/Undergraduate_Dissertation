#Remove unwanted columns

The only continent contained in this dataset is Europe meaning continent features can be dropped. Competition types are all league games and game id is not an important metric so they can also be dropped.

```python
#Continent is all europe
#Competition type is all domestic league
data = data.drop(columns=['continent', 'team_continent', 'opp_continent', 'competition_type', 'game_id'])
data
```

```python
data['opp_competition_id'].unique()
```

Check for identical columns

```python
#team_country and opp_country are identical
data['country'].equals(data['team_country'])
```

```python
data['country'].equals(data['opp_country'])
```

```python
data['opp_competition_id'].equals(data['competition_id'])
```

```python
data['team_competition_id'].equals(data['competition_id'])
```

```python
data['team_id'].equals(data['opp_opponent_id'])
```

```python
data['opponent_id'].equals(data['opp_team_id'])
```

The above shows there are several duplicate columns. These columns are dropped to avoid data redundancy.

```python
#Remove duplicates
data = data.drop(columns=['team_country', 'opp_country', 'opp_competition_id','team_competition_id', 'opp_opponent_id', 'opp_team_id', 'date', 'team_id', 'opponent_id', 'opp_competition_id', 'formation', 'opp_formation', 'competition_id'])
data.head(3)
```

```python
data.columns
```

#Final Data

Data columns are renamed to allow for shorter names

```python
data = data.rename(columns={"goal_difference": "gd", "opp_goal_difference": "opp_gd", "average_starter_age": "avg_starter_age", "opp_tm_formation": "opp_formation", "tm_formation": "formation", "matchday": "md"})
data.head(3)
```

The target variable is averaged by year to investiagte an overall trend.

```python
result = data.groupby('season')['total_starter_value'].mean()
result = pd.DataFrame(result)
result = result.reset_index()
result
```

This data is plotted

```python
result.plot(x = 'season', y='total_starter_value', kind='scatter')
```

This plot shows a positive trend from 2004 -> ~2018 followed by a reduction/plateau.

The data is prepared for plotting by matplotlib.

```python
x = result['season']
```

```python
y = result['total_starter_value']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.plot(x, y)
plt.xlabel("Year")
plt.ylabel("Starting 11 average value")
plt.xticks(x);
```
